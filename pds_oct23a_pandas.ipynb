```python
def fact(n):
    
    if n < 0:
        raise(ValueError('Boss. Wrong INput'))
    if n == 0:
        return 1
    
    return n * fact(n-1)
```

# Global and Local

```python
g1 = 11
g2 = 22

def mf1(v3):
    v4 = 444  #  v3 and v4 are in local namespace
    g1 = 111
    print(g1,g2,v3,v4)
    
print(g1,g2)
mf1(333)
print(g1,g2)
#print(v3)
#print(v4)
```

```python
g1 = 11
g2 = 22

def mf1(v3):
    global g1
    v4 = 444  #  v3 and v4 are in local namespace
    g1 = 111
    print(g1,g2,v3,v4)
    print(locals())
    
print(g1,g2)
mf1(333)
print(g1,g2)
#print(v3)
#print(v4)
```
